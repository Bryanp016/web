## Calcular Suavizacion Exponencial

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# el DataFrame se llama movil
exporta = {
    'A単o':[2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2011],
    'Exportaciones':[5501.0, 6232.7, 8118.3, 10137.00, 10449.50, 12794, 13193.00, 16036.2, 18496.90, 18709.30, 19363.50,
     16927.00, 16927.00, 5501.0]
}
                  
movil = pd.DataFrame(exporta)
# mostramos los 5 primeros registros

movil.head()

alfa = 0.1
unoalfa = 1. - alfa

for i in range(0,movil.shape[0]-1):
    movil.loc[movil.index[i+1],'SN'] = np.round(movil.iloc[i,1],1)
    
for i in range(2,movil.shape[0]):
    movil.loc[movil.index[i],'SN'] = np.round(movil.iloc[i-1,1],1)*alfa + np.round(movil.iloc[i-1,2],1)*unoalfa
    
i=i+1
p1=0
p2=np.round(movil.iloc[i-1,1],1)*alfa + np.round(movil.iloc[i-1,2],1)*unoalfa
a = movil.append({'A単o':2018,'Exportaciones':p1, 'SN':p2},ignore_index=True)
a
```
